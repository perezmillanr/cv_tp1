```python
# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
import glob
```

```python
# Cargar una imagen en modo monocromático (un canal)
PATH='./white_patch/'
img = cv.imread(PATH+'test_blue.png')

PATH_WP='./white_patch/'
PATH_CC='./coord_cromaticas/'


images_wp = [cv.imread(file) for file in glob.glob(PATH_WP+'*.png')]
images_cc = [cv.imread(file) for file in glob.glob(PATH_CC+'*.png')]

#Noto que es BGR, lo cambio a RGB
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

for index,im in enumerate(images_wp):
    images_wp[index] = cv.cvtColor(im, cv.COLOR_BGR2RGB)

for index,im in enumerate(images_cc):
    images_cc[index] = cv.cvtColor(im, cv.COLOR_BGR2RGB)


# Dimensión de la imagen
print(f'Tamaño imagen: {img.shape} pix')

#Graficamos la imagen
plt.imshow(img)
plt.show()
```

```python
def change_coord_crom(input_image:np.ndarray) -> np.ndarray: 
    '''
    Argumentos
    input_image: Imagen de entrada
    ---------------------------------------
    Retorno
    Imagen: (numpy.ndarray) en coordenadas cromaticas
    '''
    rgb=np.sum(input_image,axis=2)
    rgb = np.repeat(rgb[:, :, np.newaxis], 3, axis=2)
    img_new=np.divide(input_image, rgb, out=np.zeros_like(input_image).astype(float), where=rgb.astype(float)!=0)
    return img_new
```

```python
def wp(input_image:np.ndarray) -> np.ndarray: 
    '''
    Argumentos
    input_image: Imagen de entrada
    ---------------------------------------
    Retorno
    Imagen: (numpy.ndarray) en coordenadas cromaticas
    '''
    rgb_max=[np.amax(img[:,:,0]),
            np.amax(img[:,:,1]),
            np.amax(img[:,:,2])
            ]
    rgb_max=np.ones(input_image.shape)*rgb_max
    v255=np.ones(input_image.shape)*float(255)
    img3=(v255*input_image/rgb_max).astype(int)
    img3=np.clip(img3, 0, 255, out = img3)
    return img3.astype(int)
```

```python
img2=change_coord_crom(img)
```

```python
#Graficamos la imagen
plt.imshow(img2)
plt.show()
```

```python
#Implementacion de White patch
plt.imshow(wp(img))
plt.show()
```

```python
#Repetimos para todas las imagenes

for index,im in enumerate(images_cc):
        print(f'Imagen: {index}')
        
        print('Original')
        plt.imshow(im)
        plt.show()
        
        print('Coordenadas Cromáticas:')
        plt.imshow(change_coord_crom(im))
        plt.show()

for index,im in enumerate(images_wp):
        print(f'Imagen: {index}')
        
        print('Original')
        plt.imshow(im)
        plt.show()

        print('White-Patch:')
        plt.imshow(wp(im))
        plt.show()
```

No he notado fallas en particular en el caso de White-Patch, mas bien he notado que para la imágen 2, se ha acentuado el ruido. Entiendo que esto puede deberse a que los valores máximos de algún canal seán bajos, extendiendose hacia los 255, agrandando el ruido.
